```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df
```

This data sets consists of 3 different types of iris flowers' (Setosa, Versicolour, and Virginica) petal and sepal length.

# Histograms

```python
plt.hist(iris_df['sepal length (cm)'], color='lightgreen', edgecolor='black')
plt.title('Histogram of sepal length (cm)')
```

```python
# Distribution
sns.histplot(iris_df['sepal length (cm)'], kde=True,bins=30)
```

```python
# Plotting histograms for each feature
iris_df.hist(edgecolor='black')
plt.suptitle('Histograms of Iris Dataset Features', fontsize=16)
plt.show()
```

Here are some inferences that can be drawn from the histograms for the Iris dataset:

### **1. Sepal Length**:
- The data appears to be somewhat normally distributed, but with a slight right skew. The majority of the values are clustered between 5 and 6 cm.


### **2. Sepal Width**:
- The distribution is not normal; it has two prominent peaks, indicating a **bimodal distribution**. The data points are spread more broadly, ranging from 2 cm to 4 cm.


### **3. Petal Length**:
- This feature shows a **strong right-skew**. The majority of the data points are concentrated on the lower end (around 1 to 2 cm), with fewer data points in the higher range.


### **4. Petal Width**:
- Similar to petal length, the distribution is **right-skewed**, with most values clustered between 0.1 cm and 0.6 cm.


### **Overall Insights**:
- **Skewness**: Both petal length and petal width are right-skewed, indicating that most Iris flowers in the dataset have shorter petals.
- **Bimodal distribution**: Sepal width has a bimodal distribution, which might indicate that different species of Iris exhibit different sepal widths.


```python
iris_df.skew()
```

```python
# -------------------------------
# BEFORE LOG TRANSFORMATION
# -------------------------------
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(iris_df['sepal length (cm)'], kde=True,bins=30)
plt.title("Before Log Transformation (Sepal Length)")

# -------------------------------
# APPLY LOG TRANSFORMATION
# -------------------------------
iris_df['sepal_length_log'] = np.log(iris_df['sepal length (cm)'])

# -------------------------------
# AFTER LOG TRANSFORMATION
# -------------------------------
plt.subplot(1,2,2)
sns.histplot(iris_df['sepal_length_log'], kde=True,bins=30)
plt.title("After Log Transformation (Sepal Length - Log)")

plt.tight_layout()
plt.show()
```

The describe() function in a pandas DataFrame provides a quick overview of the statistical properties of the data.

It generates descriptive statistics for each numerical column in the DataFrame including
* Count: Number of non-null entries in each column.
* Mean: Average value of the column.
* Standard Deviation (std): Measure of the spread or dispersion of the values in the column.
* Minimum (min): The smallest value in the column.
* 25th Percentile (25%): Value below which 25% of the data falls (also known as the first quartile).
* 50th Percentile (50%): Median value of the column, below which 50% of the data falls.
* 75th Percentile (75%): Value below which 75% of the data falls (also known as the third quartile).
* Maximum (max): The largest value in the column.

```python
# Plotting boxplots for each feature
iris_df.plot(kind='box', subplots=True, layout=(1, 5), figsize=(12, 6),
             color='lightblue')
plt.suptitle('Boxplots of Iris Dataset Features', fontsize=16)
plt.show()
```
