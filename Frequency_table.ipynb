```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create sample data
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'B', 'C', 'B', 'C', 'A', 'C', 'B', 'A', 'C'],
    'Scores': [85, 90, 88, 75, 95, 88, 90, 78, 88, 80, 90, 78, 92, 85, 77],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female']
}
```

```python
# Convert the data to a DataFrame
df = pd.DataFrame(data)
df
```

Frequency means the number of times a value appears in the data. A table can quickly show us how many times each value appears.

# Frequency Table for Qualitative Data


```python
# Frequency table for 'Category'
category_freq = df['Category'].value_counts()
print("Frequency Table for Category:\n", category_freq)
```

```python
# Plot for 'Category'
sns.countplot(x='Category', data=df, palette='viridis',hue='Category')
plt.title("Frequency Distribution: Category")
plt.xlabel("Category")
plt.ylabel("Frequency")
```

```python
# Plot pie chart
plt.figure(figsize=(7,7))
plt.pie(category_freq, labels=category_freq.index, autopct='%1.2f%%',
        explode=[0.02]*len(category_freq))
plt.title("Category Distribution - Pie Chart")
plt.show()
#autopct: automatic percentage
#'%1.2f%%':show percentage on each slide with two decimal place(f-float)
```

```python
# Frequency table for 'Gender'
gender_freq = df['Gender'].value_counts()
print("Frequency Table for 'Gender':\n", gender_freq)
```

```python
# Plot pie chart
plt.figure(figsize=(7,7))
plt.pie(gender_freq, labels=gender_freq.index, autopct='%1.1f%%',
        explode=[0.02]*len(gender_freq))
plt.title("Gender Distribution - Pie Chart")
plt.show()
```

```python

# Plot for 'Gender'

sns.countplot(x='Gender', data=df, hue='Gender',palette='magma')
plt.title("Frequency Distribution: Gender")
plt.xlabel("Gender")
plt.ylabel("Frequency")
```

# Frequency Table for Numerical Data

```python
# Frequency table for 'Scores'
Scores_freq = df['Scores'].value_counts()
print("Frequency Table for 'Scores':\n", Scores_freq)
```

```python
# Plot for 'Score'
sns.countplot(x='Scores', data=df, palette='magma',hue='Scores')
plt.title("Frequency Distribution: Score")
plt.xlabel("Scores")
plt.ylabel("Frequency")
```

If the numerical data has many different values, it is easier to use intervals of values to present them in a table.

```python
bins = [70, 80, 90, 100]   # score range edges
labels = ['Low', 'Medium', 'High']

df['Score_Group'] = pd.cut(df['Scores'], bins=bins, labels=labels)
print(df[['Scores', 'Score_Group']])
```

```python

# Plotting Frequency Distribution for Score

sns.countplot(x='Score_Group', data=df, palette='cool', legend=True,
              hue='Score_Group')
plt.title("Frequency Distribution: Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create 4 equal-sized bins using qcut
df['Score_Bin'] = pd.qcut(df['Scores'], q=4)
# automatically create equal-sized bins using qcut()
print(df[['Scores', 'Score_Bin']])
```

```python
sns.countplot(x='Score_Bin', data=df)
```
