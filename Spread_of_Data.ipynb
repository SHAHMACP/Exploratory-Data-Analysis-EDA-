# Calculating variance and standard deviation using Formula

image.png

```python
# @title
#
data =[5, 7, 9, 10, 12]
mean = np.mean(data)
differences = [x - mean for x in data]
Squared_differences = [d**2 for d in differences]

data = {
    'data' :[5, 7, 9, 10, 12],
    'Differences from mean' : [x - mean for x in data],
    'Squared_differences' : [d**2 for d in differences]
}

# Converting the data into a DataFrame
df = pd.DataFrame(data)

# Display the dataset
print("Dataset:")
print(df)
```

```python
# @title
# Plot data points and mean
data = [5, 7, 9, 10, 12]

plt.figure(figsize=(10, 6))
plt.plot(data, 'o', label='Data Points')
plt.axhline(y=mean, color='r', linestyle='-', label='Mean')

# Add lines for differences between each data point and the mean
for i, value in enumerate(data):
  #enumerate(data): This function returns both the index (i) and the value (value) from the data list or array
    plt.plot([i, i], [value, mean], 'g--')

# the code draws a vertical dashed green line from the data point's value on the y-axis down (or up) to the mean value.


plt.title('Data Points and Mean')
plt.xlabel('Data Point Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

## **The population variance**

```python
# Step 4: Calculate the population variance and the population standard deviation
pop_var = sum(Squared_differences) / len(data)
pop_std = np.sqrt(pop_var)

print("Variance:", pop_var)
print("Standard Deviation:", pop_std)
```

## **The Sample variance**

```python
# Step 4*: Calculate the sample variance and the sample standard deviation
sam_var = sum(Squared_differences) / (len(data)-1)
sam_std = np.sqrt(sam_var)

print("Variance:", sam_var)
print("Standard Deviation:", sam_std)
```

# Calculating variance and standard deviation using Statistics module

## The sample variance

```python
var = stats.variance(data)
print("The sample Variance is :", var)
```

```python
stdev = stats.stdev(data)
print("Standard Deviation is:", stdev)
```

## The population variance

```python
p_var = stats.pvariance(data)
print("The Population Variance is :", p_var)
```

```python
p_stdev = stats.pstdev(data)
print("The population Standard Deviation is:", p_stdev)
```
