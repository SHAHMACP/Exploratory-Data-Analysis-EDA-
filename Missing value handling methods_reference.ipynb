### Drop Columns With Very High Missing Values (>50%)

## Filling Missing values using central tendency measures

### Handle Categorical Columns Using Mode

```python
payment_mode = df["Payment_Method"].mode()[0]
df["Payment_Method"] = df["Payment_Method"].fillna(payment_mode)
```

### Handle Numerical Columns

## Advanced Imputation Methods

### KNN imputer

> KNN Imputer fills missing values using the **average value of the k nearest samples**, based on similarity across other features.

* Example: If two customers are similar in age, income, and tenure, they are likely to have similar missing values.

> How KNN Imputer Works

1. Identify rows with missing values

2. Find k nearest neighbors using distance (usually Euclidean)

3. Compute:

- Mean (for numeric)

- Mode (implicitly via mean encoding for categorical if encoded)

4. Replace missing value

```python
knn_features = [
    'Tenure_Years',
    'Age',
    'Annual_Income',
    'Annual_Spend'
] #Choose features that logically influence customer tenure:
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[knn_features])
# KNN uses distance  features must be on the same scale.
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
imputed_data = imputer.fit_transform(scaled_data)
```

```python
df[knn_features] = scaler.inverse_transform(imputed_data)
# Inverse Scale Back
```
