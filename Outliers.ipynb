```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Dataset with outliers

```python
# Creating a dataset with outliers
data_with_outliers = {
    'Age': [25, 30, 22, 35, 28, 29, 27, 24, 26, 40],
    'Income': [50000, 60000, 55000, 70000, 58000, 65000, 62000, 59000, 53000, 1500000],  # Outlier: '1500000'
    'Work_Experience': [1, 3, 2, 10, 4, 3, 2, 3, 5, 15]
}
```

```python
# Converting the data into a DataFrame
df_with_outliers = pd.DataFrame(data_with_outliers)

# Display the dataset
print("Dataset:")
print(df_with_outliers)
```

```python
# The average income

mean_value_with_outliers = np.mean(df_with_outliers['Income'])
print("The mean Income value of the compant is", mean_value_with_outliers)
```

```python
# Scatter plot with outliers

plt.scatter(df_with_outliers['Age'], df_with_outliers['Income'])


plt.title('Scatter Plot with Outliers')
plt.xlabel('Age')
plt.ylabel('Income')
plt.axhline(y=mean_value_with_outliers, color='y')


age_limits = [0, 45]
income_limits = [40000, 1600000]
plt.xlim(age_limits)
plt.ylim(income_limits)
```

```python
# Scatter plot with outliers

plt.scatter(df_with_outliers['Work_Experience'], df_with_outliers['Income'])
plt.title('Scatter Plot with Outliers')
plt.xlabel('Work_Experience')
plt.ylabel('Income')
plt.axhline(y=mean_value_with_outliers, color='y')

Work_Experience_limits = [0, 16]
income_limits = [40000, 1600000]
plt.xlim(Work_Experience_limits)
plt.ylim(income_limits)
```

# Dataset without outliers

```python
# Creating a dataset with outliers
data = {
    'Age': [25, 30, 22, 35, 28, 29, 27, 24, 26],   # 40 removed from data_out
    'Income': [50000, 60000, 55000, 70000, 58000, 65000, 62000, 59000, 53000],  # Outlier is removed from the list
    'Work_Experience': [1, 3, 2, 10, 4, 3, 2, 3, 5]   # 15 removed from data_out
}
```

```python
# Converting the data into a DataFrame
df = pd.DataFrame(data)

# Display the dataset
print("Dataset:")
print(df)
```

```python
# The average income

mean_value = np.mean(df['Income'])
print("The mean Income value of the compant is", mean_value)
```

```python
# Scatter plot with outliers

plt.scatter(df['Age'], df['Income'], label='Data Points')
plt.title('Scatter Plot with Outliers')
plt.xlabel('Age')
plt.ylabel('Income')
plt.axhline(y=mean_value, color='r', linestyle='-')

age_limits = [20, 45]
income_limits = [40000, 1600000]
plt.xlim(age_limits)
plt.ylim(income_limits)
```
